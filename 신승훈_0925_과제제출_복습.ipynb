### 히스토그램
- 도수분포표를 시각화 한 것

```python
x = np.random.normal(5, 3, 100)
```

```python
plt.hist(x) # 히스토그램을 그릴 수 있음
```

- 구간 개수 조정가능
- bins = 수

```python
# bins 그룹 묶어서 나타냄
plt.hist(x, bins=5) # bins는 히스토그램에서 x축을 몇개의 그룹으로 묶어서 나타낼지를 결정해준다.
```

```python
plt.hist(x, bins=10) # 구간 개수를 10개의 그룹으로 조정
```

```python
plt.hist(x, bins=50) # 구간 개수를 50개의 그룹으로 조정
```

### 히트맵 

```python
data = pd.DataFrame({'math':[50,60,70],
                     'english':[60,70,80]},
                   index=['홍길동','김영희','박철수'])
```

```python
data
```

```python
sns.heatmap(data) # heatmap을 그리기 위해 데이터를 불러온다
```

```python
# df: 데이터, vmin=최솟값, vmax=최댓값, cbar=True colorbar의 유무, center=중앙값 선정, linewidths=0.5 #cell사이에 선을 집어 넣는다. cmap=blue 히트맵의 색을 설정한다 

sns.heatmap(data, annot=True, fmt='d') #데이터 값이 들어갈 수 있다
                                        # 각 cell의 값 표기 유무, 그 값의 데이터 타입 설정
                                        # 정수
```

```python
sns.heatmap(data, annot=True, fmt='.2f') # 데이터 타입 소수 둘재자리까지
```

```python
sns.heatmap(data, annot=True, fmt='.2f', linewidth=1, linecolor = 'black') # 선 두께와 색상 설정
```

```python
sns.heatmap(data, annot=True, cmap='Blues', fmt='.2f', linewidth=1, linecolor = 'black')

# cmap=blue 히트맵의 색을 설정한다
```

### 2개 이상의 그래프를 그리고 싶어요! 

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
#gridspec 이 모듈을 가지고 온다!
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) #본인이 그리고 싶은 그래프의 개수를 적으면 됩니다.
plt.figure(figsize=(15,10))
plt.subplots_adjust(wspace=0.5, hspace=0.3) #현재 배치되어 있는 subplot들의 위치를 직접 조정한다
                                            # wspace 가로 간격, hspace 세로 간격

    
#서브플롯에 필요한 컬럼을 가져온다!
mpg_features = ['manufacturer','model','category','hwy']

for idx, feature in enumerate(mpg_features): # 인덱스(index)와 원소를 동시에 접근하면서 루프를 돌릴 수 있다.
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=feature,      # feature calss
                 data=mpg,        # data는 mpg 데이터
                 hue='drv',       # drv 컬럼의 값별로 count
                 palette='pastel',# 색 구성
                 ax=ax)           # 그래프위 위치 결정
    ax.set_title(f'{feature} Distribution')
```

```python
#gridspec 이 모듈을 가지고 온다!
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) #본인이 그리고 싶은 그래프의 개수를 적으면 됩니다.
plt.figure(figsize=(15,10))
plt.subplots_adjust(wspace=0.5, hspace=0.3) #현재 배치되어 있는 subplot들의 위치를 직접 조정한다
                                            # wspace 가로 간격, hspace 세로 간격

    
#서브플롯에 필요한 컬럼을 가져온다!
mpg_features = ['manufacturer','model','category','hwy']

for idx, feature in enumerate(mpg_features): # 인덱스(index)와 원소를 동시에 접근하면서 루프를 돌릴 수 있다.
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=feature,      # feature calss
                 data=mpg,        # data는 mpg 데이터
                 hue='drv',       # drv 컬럼의 값별로 count
                 palette='pastel',# 색 구성
                 ax=ax,           # 그래프위 위치 결정
                 dodge=False)    # 누적 그래프 (hue 변수 분리하지 않고 위로 쌓아 올린다)
    ax.set_title(f'{feature} Distribution')
```
