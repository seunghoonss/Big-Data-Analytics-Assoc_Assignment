### Q5. 제조사(manufacturer)에 따른 자동차의 복합 연비(cty+hwy/2)가 어떻게 다른지 비교해보는 요약통계

```python
# 파생변수 만들기!
# lambda 식을 이용한 파생변수 만들기!
df_nc5 = df.query('manufacturer == ["audi", "chevrolet"]' or 'cty>=1' or 'hwy>=1')

df_nc5_1 = df_nc5.assign(total = lambda x: x['cty']+ x['hwy'], # 합
           mean = lambda x: x['total']/2).sort_values(['cty', 'hwy', 'mean'], ascending = [False, False, False]) # 평균
df_nc5_1.iloc[:, 0:13].drop(columns = ['model', 'year', 'displ', 'cyl', 'trans', 'drv', 'fl', 'category', 'total'])

# 새로운 df_nc5_1 파생변수를 만들기 
# 위의 제조사의 도시연비와 고속도로연비의 합과 평균값을 계산한다.
# cty, hwy, mean 모두 내림차순 정렬을 해준다.
```

```python
# 집단별로 요약하는 Groupby
# 추출한 데이터를 이용해 두 제조사에 대한 평균 복합 연비값(mean_gasmileage)을 계산해라!


df_nc5_1.groupby('manufacturer').agg(mean_cty = ('cty', 'mean'),
                                    mean_hwy = ('hwy', 'mean'),
                                   mean_gasmileage = ('mean', 'mean'))

# adui 평균 복합 연비: 22.03 이다.
# chevrolet 평균 복합 연비: 18.45 이다.
# audi 가 chevrolet 보다 평균 복합 연비가 약 3.58더 좋은 것을 볼 수 있다.
```

### Q45. 차량등급(midsize)에 따른 자동차의 복합 연비(cty+hwy/2)가 어떻게 다른지 비교해보는 요약통계

```python
# 파생변수 만들기!
# lambda 식을 이용한 파생변수 만들기!
df_nc45 = df.query('category == ["compact", "midsize"]' or 'cty>=1' or 'hwy>=1')

df_nc45_1 = df_nc45.assign(total = lambda x: x['cty']+ x['hwy'], # 합
           mean = lambda x: x['total']/2).sort_values(['cty', 'hwy', 'mean'], ascending = [False, False, False]) # 평균
df_nc45_1.iloc[:, 0:13].drop(columns = ['manufacturer', 'model', 'displ', 'year', 'cyl', 'trans', 'drv', 'fl', 'total'])

# 새로운 df_nc45 파생변수를 만들기 
# 위의 차량등급의 도시연비와 고속도로연비의 합과 평균값을 계산한다.
# cty, hwy, mean 모두 내림차순 정렬을 해준다.
```

```python
# 집단별로 요약하는 Groupby
# 추출한 데이터를 이용해 두 차량등급에 대한 평균 복합 연비값(mean_gasmileage)을 계산해라!


df_nc45_1.groupby('category').agg(mean_cty = ('cty', 'mean'),
                                    mean_hwy = ('hwy', 'mean'),
                                   mean_gasmileage = ('mean', 'mean'))

# compact 차량등급 차량의 평균 복합 연비: 24.21 이다.
# midsize 차량등급 차량의 평균 복합 연비: 23.02 이다.
# compact가 midsize 차량등급보다 평균 복합 연비가 약 1.19 더 좋은 것을 볼 수 있다.
```
