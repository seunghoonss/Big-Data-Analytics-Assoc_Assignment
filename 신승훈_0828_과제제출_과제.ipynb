## 과제2 

```python
import seaborn as sns # searborn 라이브러리 불러온다
```

```python
sns.boxplot(data = mpg, y = 'cty') # cty 컬럼 이상치 살펴본다.
```

```python
# 1사분위수, 3사분위수 값
# quantile() 분위수를 구할 수 있다!

qut25 = mpg['cty'].quantile(.25) # 하위 25%
qut75 = mpg['cty'].quantile(.75) # 하위 75%
```

```python
display(qut25) # 하위 25%: 14.0
display(qut75) # 상위 25%: 19.0
```

```python
# IQR 구하기
# 1사분위수와 3사분위수의 거리를 IQR(Inter Quantile range, 사분위범위) 구한다!

iqr = qut75 - qut25 # IQR 구하기
iqr
```

```python
# 상한, 하한을 구한다! 

qut25 - 1.5 * iqr # 하한
```

```python
qut75 + 1.5 * iqr # 상한
```

## 극단치를 결측 처리하기!!! 

```python
import numpy as np # numpy 라이브러리 불러온다
mpg['cty'] = np.where((mpg['cty']<6.5)|(mpg['hwy']>26.5),np.nan, mpg['hwy']) # 조건에 만족하는 cty 값
```

```python
mpg['cty'] # 극단치 결측을 Nan 값으로 처리한다!
```

```python
mpg['cty'].isna().sum() # 이상치가 Nan으로 처리된 결측치가 69개가 나온것을 확인 할 수 있다.
```

```python
mpg.groupby('drv').agg(mean_hwy=('cty','mean')) # 결측지를 제거 안하고 분석
```

```python
mpg.dropna(subset=['cty']).groupby('drv').agg(mean_hwy=('cty','mean')) # 결측치를 제거 하고 분석
```
