# matplot
- 라인 플롯, 바 차트, 파이차트, 히스토그램, Box plot, scatter plot 등 다양한 차트와 프롯 스타일을 지원

```python
# 산점도 plt.scatter([])
plt.scatter([1, 2, 3], [4, 5, 6])
```

```python
# 바차트 plt.bar([])
plt.bar([1, 2, 3], [4, 5, 6])
```

```python
# 파이차트 plt.bar([])
plt.figure(figsize=(1,1))
plt.pie([1, 2, 3], [1, 1, 1])
```

```python
# 히스토그램 차트 plt.bar([])
plt.hist([1, 2])
```

```python
# fill_demo plt.subplots()
x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

fig, ax = plt.subplots()

ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
plt.show()
```

```python
# line_demo_dash_contorl

x = np.linspace(0, 10, 500)
dashes = [10, 5, 100, 5]  # 10 points on, 5 off, 100 on, 5 off

fig, ax = plt.subplots()
line1, = ax.plot(x, np.sin(x), '--', linewidth=2,
                 label='Dashes set retroactively')
line1.set_dashes(dashes)

line2, = ax.plot(x, -1 * np.sin(x), dashes=[30, 5, 10, 5],
                 label='Dashes set proactively')

ax.legend(loc='lower right')
plt.show()
```

```python
# errorbar_demo
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

fig, ax = plt.subplots()
ax.errorbar(x, y, xerr=0.2, yerr=0.4)
plt.show()
```

```python
# streamplot_demo_features

Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

fig0, ax0 = plt.subplots()
strm = ax0.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
fig0.colorbar(strm.lines)

fig1, (ax1, ax2) = plt.subplots(ncols=2)
ax1.streamplot(X, Y, U, V, density=[0.5, 1])

lw = 5*speed / speed.max()
ax2.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)

plt.show()
```

```python
# Compute pie slices: polar_bar_demo

N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)
```

```python
# Compute areas and colors # poloar_scatter_demo
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

ax = plt.subplot(111, projection='polar')
c = ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75)

plt.show()
```

## seaborn
- Seaborn은 Matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키치이다.
- Relational, Categorical, Distribution, Regression, Matrix, Multi-plot grid

```python
import seaborn as sns
```

```python
# seaborn에서 제공하는 flights 데이터 셋을 사용

flights = sns.load_dataset('flights')
plt.figure(figsize=(12, 3))
flights # flights 데이터셋 Dataframe
```

```python
# Barplot

sns.barplot(data=flights, x = 'year', y='passengers')
# 각가의 값에 검정색 막대기는 신뢰구간을 나타낸다
```

```python
# violinplot
sns.violinplot(data=flights, x = 'year', y='passengers') # 데이터의 4분위수 기준으로 상세한 분포를 도식화
```

```python
# swarmplot
sns.swarmplot(data=flights, x = 'year', y='passengers') # 데이터의 분포를 확인한다
```

```python
# Boxplot
sns.boxplot(data=flights, x="year", y="passengers")
```

```python
# Lineplot
sns.lineplot(data=flights, x="year", y="passengers") # 상위 점을 하나의선으로 이은 형태 # 주변의 옅은 파란색은 데이터의 신뢰구간
```

```python
# rugplot
sns.rugplot(flights["passengers"]) # 선분(rug)으로 나타내어 실제 데이터들의 위치를 보여준다
```

```python
sns.kdeplot(flights["passengers"]) # 커널이란 함수를 겹치는 방법으로 히스토그램보다 부드러운 형태의 분포 곡선을 보여준다.
```

```python
# Displot
sns.distplot(flights["passengers"]) # displot은 하나의 데이터에 대해 분포를 확인할 때 사용한다.
```

```python
# 다차원 데이터
# 2차원 실수형 데이터

sns.jointplot(flights["passengers"], flights['year'], kind="scatter")
sns.jointplot(flights["passengers"], flights['year'], kind="kde")

# joinplot 명령은 스캐더 플롯뿐 아니라 차트의 가장자리에 각 변수의 히스토그램 그린다.
# kind = 'scatter' 스캐터 플롯
# kind = 'kde' 커널 밀도 히스토그램
```

```python
# stripplot 모든 데이터를 점으로 그려준다.

sns.stripplot(flights["passengers"], flights['year'],  jitter=True)
# jitter=True 가로축상의 위치를 무작위로 바꾸어서 데이터의 수가 많을 경웨 겹치지 않도록 한다.
```

### 범례 위치 바꾸기
- 범례의 위치를 바꿀 수 있는 인자
- left
- center
- right
- upper left
- upper center
- upper right
- lower left
- lower center
- lower right

```python
plt.scatter([1,2,3],[4,5,6], label ='x')
plt.scatter([-1,-2,-3],[4,5,6], label ='y')
plt.legend(loc ='center')#라벨의 범례 출력
```

### 마커 설정도 가능하다!
- .(점), 'o', 's'사각형, '*'별, 'D'다이아몬드 등등~ 
- 구글링해서 찾아보기!!

- markersize, marker

```python
plt.plot([1,2,3],[4,5,6],marker = '*',markersize=10,color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 's',markersize=10,color = 'blue', label ='line y')
plt.legend(loc='center')
```

```python
# marker 마커 설정: o, s
plt.plot([1,2,3],[4,5,6],marker = 'o',markersize=10,color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 's',markersize=10,color = 'blue', label ='line y')
plt.legend(loc='center')
```

```python
# marker 마커 설정: *, D
plt.plot([1,2,3],[4,5,6],marker = '*',markersize=10,color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 'D',markersize=10,color = 'blue', label ='line y')
plt.legend(loc='center')
```
