### Q2. 데이터 병합하기 

```python
mpg1 = mpg.iloc[:, :4] # mpg1 dataframe 생성
display(mpg1)
```

```python
mpg2 = mpg.iloc[:, 2:] # mpg2 dataframe 생성 # 교지합이 2개 이상인 경우로 설정!
display(mpg2)
```

```python
# 병합 진행
df = pd.concat([mpg1, mpg2], axis = 0) # 행을 기준으로 함
```

```python
# 병합 진행
pd.concat([mpg1, mpg2], axis = 1) # 열을 기준으로 함
```

```python
pd.merge(mpg1, mpg2, how='inner', on=['displ', 'year']) # 정밀 교집합만 displ, year를 기준으로 모두 다 같아야 한다!
```
