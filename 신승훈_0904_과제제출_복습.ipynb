### 시계열 데이터 프레임 만들기
- datetime 판다스에서 제공하는 시계열 데이터 형식
- 시계열 데이터는 Dtype이 꼭 datetime으로 변경해야 한다!
- 이유는 datetime 문법이 따로 있고, 이 문법을 적용시키려면 Dtype이 꼭 datetime으로 설정되어 있어야 한다.

```python
df = pd.read_csv('country_timeseries.csv') # country_timeseries csv파일을 불러온다
```

```python
df
```

```python
df.info() #문자열 데이터 #Dypte이 object이다.
```

```python
df['Date'] = pd.to_datetime(df['Date']) #datetime으로 변경해서 시계열 형식으로 맞춤
                                        # datetime형으로 타입 변환하기: .to_dataetime()
```

```python
df.info() # Dtype이 datetime으로 변경된 것을 볼 수 있다!
```

```python
df['Date'] # 날짜가 슬래시로 구분되어 적혀있다.
```

```python
df
```

```python
df[df['Date'].str.contains('2015')] #datetime형태로 바뀐 후엔 문자열만 사용할 수 있는 .str함수는 사용할 수 없다.
```
